```python
#Lab 1.2 — Closure with Changing State
def counter():
   count = 0
   def inc():
    nonlocal count
    count += 1
    print("count =", count)
   return inc

c = counter()
c()
c()
c()

#Lab 1.3 — Prove Where the Data Lives
print(c.__closure__)
print(c.__closure__[0].cell_contents)
```

```python
#PART 2 — Decorators
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def greet():
  print("Hello!")


greet()
```

1. The wrapper function runs first, specifically executing the code before the original function call.
2. No, the original code of greet() remains untouched.
3. The wrapper function defined inside my_decorator added the extra behavior.

```python
#Lab 2.2 — Decorator vs Wrapper
def demo_decorator(func):
   print("DECORATOR runs (definition time)")
   def wrapper():
     print("WRAPPER start")
     func()
     print("WRAPPER end")
   return wrapper

@demo_decorator
def say_hi():
   print("HI")
   say_hi()
```

1. The decorator runs exactly once at the time the function is defined.
2. The wrapper runs every single time the decorated function is called.
3. The original function only runs if the wrapper explicitly calls it.


```python
#Lab 2.3 — Inspect the Decorated Function
print(say_hi)
print(say_hi.__closure__)
```

1. After decoration, the function name (e.g., say_hi) now refers to the wrapper function.
2. The original function is stored securely within the wrapper’s closure.

```python
#PART 3 — Decorator with Arguments
def log_calls(func):
    def wrapper(*args):
      print("Calling with args:", args)
      return func(*args)
    return wrapper

@log_calls
def add(a, b):
    return a + b

print(add(3, 4))
print(add(10, 20))
```

1. Logging is added inside the wrapper function before the original function is called.
2. No, the logic of add remains a + b; it is simply intercepted by the decorator.
3. We use *args to allow the decorator to intercept and pass any number of positional arguments to the original function dynamically.

```python
#Lab 4.1 — Closure = Data Memory
def make_multiplier(n):
    def multiply(x):
       return x * n
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))
print(triple(5))
```
